### Loading the datasets

```python
#Loading the dataset
import pandas as pd
movies=pd.read_csv('/Users/devanshiverma/Desktop/UChicago/02-Quarter/ml-100k/u.data',sep='\t',names=["user_id", "movie_id", "rating", "timestamp"],encoding='latin-1')
users=pd.read_csv('/Users/devanshiverma/Desktop/UChicago/02-Quarter/ml-100k/u.user',sep='|',names=["user_id","age","gender","occupation", "zip_code"],encoding='latin1')
items=pd.read_csv('/Users/devanshiverma/Desktop/UChicago/02-Quarter/ml-100k/u.item',sep='|',names=["movie_id", "title", 'release_date', 'video_release_date', 'imdb_url',
                                            "unknown","Action","Adventure","Animation","Children",
                                            "Comedy","Crime","Documentary","Drama","Fantasy","Noir","Horror",
                                            "Musical","Mystery","Romance","SciFi","Thriller"
                                            ,"War","Western"],encoding='utf-8')
```

```python
print("Shape of the movie dataset is{}".format(movies.shape))
movies.head()
```

```python
print("Shape of the users dataset is{}".format(users.shape))
users.head()
```

```python
print("Shape of the items dataset is{}".format(items.shape))
items.head()
```

```python
df1=movies.merge(items,how="left",left_on="movie_id",right_on="movie_id")
df=df1.merge(users,how="left",left_on="user_id",right_on="user_id")
df.head()
```
